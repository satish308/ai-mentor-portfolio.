“Explain DSA preparation for a campus placement interview”
Data Structures and Algorithms preparation for campus placements is essentially about building pattern recognition. Most interview questions are variations of a few dozen core patterns, so your goal isn't to solve hundreds of random problems but to master those patterns until you can recognize them quickly under pressure.
How to think about your preparation
Campus interviews typically have 2-3 coding rounds, each lasting 45-60 minutes with 1-2 problems. Companies like product-based startups, FAANG, and high-paying firms (Uber, Atlassian, Adobe) lean heavily on DSA, while service companies (TCS, Infosys, Wipro) ask easier problems but cover more breadth including basic OOP and SQL. Know which bucket your target companies fall into so you calibrate effort accordingly.
A realistic timeline is 3-4 months of consistent practice if you're starting from scratch, or 6-8 weeks for a refresher if you already know the basics. Aim for 2-3 problems daily rather than 10 problems on weekends.
The core topics, roughly in order of importance
Start with arrays and strings since they appear in almost every interview. Master the sliding window, two-pointer, and prefix sum patterns. Problems like "longest substring without repeating characters," "container with most water," and "subarray sum equals k" are foundational.
Hashing comes next because it unlocks O(n) solutions for many problems. Learn when to use a hash map versus a hash set, and practice problems involving frequency counting and complement-finding.
Recursion and backtracking are conceptually harder but essential. Work through subsets, permutations, combinations, N-Queens, and Sudoku solver. Understanding the recursion tree visually helps enormously.
Linked lists show up often because they test pointer manipulation. Focus on reversal, cycle detection (Floyd's algorithm), merging sorted lists, and finding the middle node.
Trees and binary search trees are interview favorites. Master all four traversals (recursive and iterative), level-order using BFS, lowest common ancestor, diameter, and tree construction from traversals.
Stacks and queues seem simple but have clever applications: next greater element, valid parentheses, monotonic stack problems, and implementing one using the other.
Heaps and priority queues are essential for "top K" problems, merge K sorted lists, and median in a stream.
Graphs include BFS, DFS, topological sort, Dijkstra's, and union-find. For campus interviews, you mostly need BFS/DFS on grids and basic graph traversal.
Dynamic programming is the most intimidating topic. Start with 1D DP (climbing stairs, house robber, longest increasing subsequence), then 2D DP (unique paths, edit distance, longest common subsequence), then knapsack variants. Don't skip this—it's a frequent differentiator in interviews.
Greedy algorithms and binary search round things out. Binary search on answer (not just sorted arrays) is a powerful technique worth practicing.
A practical study approach
Pick one resource and stick with it rather than jumping around. Striver's SDE Sheet, NeetCode 150, or the Blind 75 are all solid curated lists. Don't try to do all three.
For each problem, give yourself 20-30 minutes to attempt it. If stuck, read the editorial, understand it deeply, then re-implement without looking. The next day, attempt it again from scratch. Spaced repetition matters more than volume.
Maintain a simple notebook (digital or paper) where you write down the pattern for each problem you solve. After 100 problems, you should be able to look at any new problem and say "this is a sliding window variant" or "this is 0/1 knapsack." That recognition is the real skill.

Weeks 11-13: Dynamic programming. Three full weeks because DP is the differentiator. 1D problems first, then 2D, then knapsack patterns, then string DP. If you can solve LIS, edit distance, and 0/1 knapsack confidently, you're ahead of most candidates.
Week 14: Greedy and binary search on answer. Often underestimated. Binary search on answer is a sleeper-hit pattern.
Weeks 15-16: Revision, mock interviews, company-specific prep. This is non-negotiable.
The practice method that actually works
Stop solving problems in isolation. Here's what I drill into my students:
For every problem, give yourself 25 minutes max. If you're stuck, read the editorial, understand it completely, then close it and re-implement from scratch. Three days later, attempt it again cold. If you still struggle, you don't actually know it—mark it for weekly review.
Maintain a pattern journal. After each problem, write one line: "This was a sliding window variant because we needed to find the longest window satisfying X." After 150 problems, you'll have built genuine pattern recognition. When you see a new problem, your brain will pattern-match in seconds. That's the actual skill interviewers test.
Solve on paper or in a plain text editor at least three times a week. Your IDE is lying to you about your real ability. In an interview, there's no autocomplete, no syntax highlighting, no instant runs.
What separates offers from rejections
I've sat in debriefs with hiring managers from Google, Amazon, and Microsoft. Here's what they actually look for, and it's almost never just "did they solve it":
Did the candidate clarify the problem before coding? Students who immediately start coding lose points even if they solve it.
Did they think out loud? Silent problem-solving terrifies interviewers because they can't follow your thinking, and they're trained to optimistically assume you're stuck.
Did they discuss edge cases unprompted? Empty input, single element, duplicates, negative numbers, integer overflow—mention these before the interviewer asks.
Did they analyze complexity correctly and honestly? Don't say O(n) when it's O(n log n). They will catch it and trust collapses.
Did they test their code with examples? Walking through your code with a sample input shows engineering maturity.
A candidate who solves only half the problem but does all of the above often beats one who solves it fully in silence.

The first three mocks will be humbling—you'll freeze, forget basic syntax, mumble. That's exactly why you need them now, not the week before TCS Codevita.
Treat every mock like a real interview. Camera on. Dressed properly. No notes. Speak aloud the entire time.

Blind 75 is fine if you're short on time. Don't switch sheets midway—it's just procrastination dressed up as optimization.
